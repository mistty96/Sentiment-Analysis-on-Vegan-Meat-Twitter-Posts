```python
import snscrape.modules.twitter as stwitter
import pandas as pd
import numpy as np
import pickle as pkl 
```

Extract Data for 2020 

```python

# Creating list to append tweet data to
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(stwitter.TwitterSearchScraper('plant based meat vegan meat lang:en since:2020-01-01 until:2021-01-01').get_items()):

    tweets_list.append([tweet.date, tweet.id, tweet.url, tweet.content, tweet.user.username, tweet.likeCount])

# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['datetime', 'id','url' ,'content', 'username', 'like_count'])
tweets_df.to_csv('tweets_2020_v2.csv', sep=',', index=False)
```

Acquire data for 2021 

```python
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(stwitter.TwitterSearchScraper('plant based meat vegan meat lang:en since:2021-01-01 until:2022-01-01').get_items()):
    #if i>maxTweets:
        #break
    tweets_list2.append([tweet.date, tweet.id, tweet.url, tweet.content, tweet.user.username, tweet.likeCount])

# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['datetime', 'id','url' ,'content', 'username', 'like_count'])

# Export dataframe into a CSV
tweets_df2.to_csv('tweets_2021_v2.csv', sep=',', index=False)
```
